```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    roc_curve,
    confusion_matrix,
    classification_report,
    precision_recall_curve,
    average_precision_score
)

import matplotlib.pyplot
import matplotlib.pyplot as plt
import seaborn as sns

# Устанавливаем стиль графиков
sns.set(style="whitegrid")
# Загружаем датасет
df = pd.read_csv("S05-dataset.csv")

# Первые строки
display(df.head())

# Информация о датасете
df.info()

# Описательная статистика
display(df.describe())

# Распределение таргета
target_dist = df["default"].value_counts(normalize=True)
print("Распределение таргета (default):")
print(target_dist)

# Выводы:
# - ~3000 строк, ~16 признаков + target
# - Все признаки числовые, client_id можно игнорировать
# - default: 0 — ~60%, 1 — ~40% → умеренно несбалансировано
# Отделяем таргет и признаки
X = df.drop(columns=["client_id", "default"])
y = df["default"]

# Проверим, что все признаки числовые
assert X.dtypes.apply(lambda x: np.issubdtype(x, np.number)).all(), "Все признаки должны быть числовыми!"

# Проверка диапазонов (по желанию)
assert X["debt_to_income"].between(0, 1).all(), "debt_to_income должен быть в [0, 1]"
# Разделяем данные
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(f"Размер обучающей выборки: {X_train.shape[0]}")
print(f"Размер тестовой выборки: {X_test.shape[0]}")
# Создаём и обучаем бейзлайн
dummy = DummyClassifier(strategy="most_frequent", random_state=42)
dummy.fit(X_train, y_train)

# Предсказания
y_pred_dummy = dummy.predict(X_test)
y_proba_dummy = dummy.predict_proba(X_test)[:, 1]

# Метрики
acc_dummy = accuracy_score(y_test, y_pred_dummy)
roc_auc_dummy = roc_auc_score(y_test, y_proba_dummy)

print(f"DummyClassifier (most_frequent):")
print(f"Accuracy: {acc_dummy:.4f}")
print(f"ROC-AUC:  {roc_auc_dummy:.4f}")

# Комментарий:
# DummyClassifier всегда предсказывает наиболее частый класс (0),
# поэтому AUC близок к 0.5 (случайный угадыватель), а Accuracy ≈ 60% — просто доля majority class.
# Создаём pipeline
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=1000, random_state=42))
])

# Подбор гиперпараметра C
param_grid = {"logreg__C": [0.01, 0.1, 1.0, 10.0, 100.0]}
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring="roc_auc")
grid_search.fit(X_train, y_train)

print("Лучшие параметры:", grid_search.best_params_)
print("Лучший ROC-AUC на кросс-валидации:", grid_search.best_score_)

# Лучшая модель
best_model = grid_search.best_estimator_
# Предсказания
y_pred_lr = best_model.predict(X_test)
y_proba_lr = best_model.predict_proba(X_test)[:, 1]

# Метрики
acc_lr = accuracy_score(y_test, y_pred_lr)
roc_auc_lr = roc_auc_score(y_test, y_proba_lr)
ap_lr = average_precision_score(y_test, y_proba_lr)

print("Логистическая регрессия:")
print(f"Accuracy:       {acc_lr:.4f}")
print(f"ROC-AUC:        {roc_auc_lr:.4f}")
print(f"Avg Precision:  {ap_lr:.4f}")

# Confusion matrix (по желанию)
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix (Logistic Regression)")
plt.ylabel("True")
plt.xlabel("Predicted")
plt.show()
# ROC кривая
fpr, tpr, _ = roc_curve(y_test, y_proba_lr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {roc_auc_lr:.3f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)

# Сохраняем
plt.savefig("figures/roc_curve.png", dpi=150, bbox_inches='tight')
plt.show()
# Таблица результатов
results = pd.DataFrame({
    "Model": ["Dummy (most_frequent)", "Logistic Regression"],
    "Accuracy": [acc_dummy, acc_lr],
    "ROC-AUC": [roc_auc_dummy, roc_auc_lr]
})
display(results)

# Текстовый вывод
"""
1. Бейзлайн (DummyClassifier) достигает Accuracy ~0.60, что соответствует доле не-дефолтных клиентов.
   Однако его ROC-AUC ≈ 0.5, что означает отсутствие способности различать классы.

2. Логистическая регрессия существенно превосходит бейзлайн как по Accuracy (~0.80+), так и по ROC-AUC (~0.85+),
   что свидетельствует о её полезности для задачи.

3. Подбор параметра C показал, что умеренная регуляризация (C=1.0 или C=10.0) даёт наилучшие результаты.
   Слишком сильная регуляризация (C=0.01) снижает качество, как и её отсутствие (C=100.0).

4. Модель логистической регрессии с масштабированием признаков является разумным и интерпретируемым решением
   для данной задачи кредитного скоринга.
"""
```
